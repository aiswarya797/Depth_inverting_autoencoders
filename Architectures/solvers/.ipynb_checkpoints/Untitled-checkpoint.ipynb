```python
child_counter = 0
for child in net.children():
    if child_counter == 1:
        #print("child ",child_counter," was frozen")
        for name, param in child.named_parameters():
            if name =='Lin_1.weight' or name =='Lin_1.bias' or name =='Lin_2.weight'  or name =='Lin_2.bias':
                param.requires_grad = False
                print(name, 'is FROZEN')
                init.kaiming_normal(param.weight,  nonlinearity='relu')
            else:
                print(name, 'is NOT frozen')
    child_counter += 1
   
```
